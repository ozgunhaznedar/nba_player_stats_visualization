```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_NBA_champions"
```

```python
dfs = pd.read_html(url)
```

```python
df = dfs[1].copy()
df.drop(0,inplace = True)
df.drop(4, inplace = True)
df.reset_index(drop=True , inplace = True)
df.head()
```

```python
df["champ_conf"] = df.Result.apply(lambda result : "W" if result.split("–")[0]  > result.split("–")[-1] else "E")
df["champ"] = df.apply(lambda row : row["Western champion"] if row.champ_conf == "W" else row["Eastern champion"]  ,axis = 1 )
df["champ"] = df.champ.apply(lambda c : c.split("(")[0].strip())
df["Year"] = df.Year.apply(lambda y: y[:4])
df["Year"] = df.Year.astype('int64')
df = df[df.Year >= 1997].reset_index(drop=True)      
df
```

```python
df = df.groupby("champ").Year.count().reset_index().sort_values(by="Year" , ascending = False).reset_index(drop=True)
df.to_csv("champs.csv")
df
```
